```python
#import all the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'SalesFINAL.csv'
inventory_data = pd.read_csv(file_path)
Invoice = pd.read_csv('InvoicePurchases.csv')
FinalPurchase = pd.read_csv('FinalPurchase.csv')
```

```python
# Display basic information about the dataset
print("Dataset Information:")
print(inventory_data.info())
```

```python
# Display the first few rows of the dataset
print("\nFirst Few Rows of the Dataset:")
inventory_data.head()
```

```python
# Summary statistics of numerical columns
print("\nSummary Statistics:")
inventory_data.describe()
```

```python
# Check for missing values
print("\nMissing Values:")
print(inventory_data.isnull().sum())
```

```python
# Distribution of SalesQuantity using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(inventory_data['SalesQuantity'], bins=20, kde=True)
plt.title('Distribution of Sales Quantity')
plt.xlabel('Sales Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
# SalesPrice vs. SalesQuantity scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SalesQuantity', y='SalesPrice', data=inventory_data)
plt.title('SalesPrice vs. SalesQuantity')
plt.xlabel('Sales Quantity')
plt.ylabel('Sales Price')
plt.show()
```

```python
# Sales by Store using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Store', y='SalesDollars', data=inventory_data)
plt.title('Total Sales by Store')
plt.xlabel('Store')
plt.ylabel('Total Sales Dollars')
plt.show()
```

```python
# Box plot of SalesDollars by Brand
plt.figure(figsize=(14, 8))
sns.boxplot(x='Brand', y='SalesDollars', data=inventory_data, palette='pastel')
plt.title('SalesDollars Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Sales Dollars')
plt.show()
```

```python
# Time series plot of SalesDollars over time
inventory_data['SalesDate'] = pd.to_datetime(inventory_data['SalesDate'])
plt.figure(figsize=(16, 8))
sns.lineplot(x='SalesDate', y='SalesDollars', data=inventory_data, color='green')
plt.title('SalesDollars Over Time')
plt.xlabel('Sales Date')
plt.ylabel('Sales Dollars')
plt.show()
```

```python
plt.figure(figsize=(3,3))
inventory_data.VendorName.value_counts()[0:5]
mylabels=["MARTIGNETTI COMPANIES","ULTRA BEVERAGE COMPANY LLP" ,'M S WALKER INC','PERFECTA WINES','E & J GALLO WINERY']


plt.pie(inventory_data.VendorName.value_counts()[0:5],labels=mylabels ,autopct='%0.1f%%',radius=1.25,
        textprops={'size':10,},shadow=True)

plt.title('Top 5 vendors\n')
plt.show()
```

```python
plt.figure(figsize=(15,5))
vc = inventory_data.groupby(['Description'])['SalesPrice'].max().sort_values(ascending=False)[:10]

g = sns.barplot(x= vc.index , y = vc.values , data = inventory_data)

for i in range(10):
    value = vc[i]
    g.text(y = value -2, x= i+0.25, s= value, color='black', ha = 'center', fontsize= 10)

plt.title ('Top 10 Brands with heighest Price tags')
plt.xlabel('Brands')
plt.xticks(rotation= 45)
plt.show()    

# As we can see "Glan grant scotch" tops the list followed by "PEL Tequilla".
```

```python
plt.figure(figsize=(12,5))

vc = Invoice.groupby(['VendorName'])['Quantity'].max().sort_values(ascending=False)[:10]
g = sns.barplot(x=vc.index, y= vc.values, data = Invoice, palette='magma')

for i in range(10):
    value = vc[i]
    g.text(y= value-2 , x= i, s= value, color='black', ha = 'center', fontsize =10)

plt.title('Vendors vs Quantity')
plt.xlabel('Vendor Names')
plt.xticks(rotation= 35)
plt.show()    

# Here, "Diago North America inc" purchased maximum quantities followed by "Martigentti companies".
```

```python
plt.figure(figsize=(15,5))
vc = FinalPurchase.groupby(['Description'])['PurchasePrice'].max().sort_values(ascending=False)[:10]
g= sns.barplot(x=vc.index, y= vc.values , data = FinalPurchase)

for i in range(10):
    value = vc[i]
    g.text(x=i , y= value , s = value , ha='center', color='black', fontsize=10)

plt.title('Top 10 Brands with heighest Purchase Price')
plt.xlabel('Brands')
plt.xticks(rotation=35)
plt.show()    

# "PEL Tequilla" has heighest Purchase Price followed by "G. Pride".
```
